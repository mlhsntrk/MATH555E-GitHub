```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd
```

```python
userdata = pd.read_csv('./whotalkstowhom.csv')
userdata.columns = ['user1', 'user2']
userdata.head(10)
```

```python
graph0 = nx.from_pandas_edgelist(userdata, source = 'user1', target = 'user2', create_using = nx.DiGraph)
print(graph0.number_of_nodes())
print(graph0.number_of_edges())
```

```python
figure(figsize = (40,32))
nx.draw_networkx(graph0, with_labels = True, node_size = 50)
```

```python
centrality = nx.degree_centrality(graph0)
centralusers = sorted(((f"{c:0.3f}", v) for v, c in centrality.items()), reverse = True)

for i in range(10):
    print(centralusers[i])
```

```python
centrality = nx.closeness_centrality(graph0.reverse(), wf_improved = True)
centralusers = sorted(((f"{c:0.3f}", v) for v, c in centrality.items()), reverse = True)

for i in range(10):
    print(centralusers[i])
```

```python
centrality = nx.betweenness_centrality(graph0)
centralusers = sorted(((f"{c:0.3f}", v) for v, c in centrality.items()), reverse = True)

for i in range(10):
    print(centralusers[i])
```
